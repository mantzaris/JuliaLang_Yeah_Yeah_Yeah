```julia
using Plots, Distributions
```

```julia
er = Normal(0,5)
```

```julia
x = 1:100
y = collect( x )
y_er = y .+ rand( er, 100 )
```

```julia
plot( x , y , label="no noise" )
plot!( x , y_er , seriestype=:scatter , label="with noise" )
```

```julia
function circle( r , angle ) 
    x = r * cos( angle )
    y = r * sin( angle )
    return [x,y]
end
```

```julia
circle( 1 , 2pi/8 )
```

```julia
angles = 0:0.05:2pi
```

```julia
coords = circle.( 2 , angles)
coords = reduce(hcat,coords)'
```

```julia
plot( coords[:,1] , coords[:,2] )
```

```julia
coords1 = reduce(hcat, circle.( 2 , 0:0.05:2pi ) )'
coords2 = reduce(hcat, circle.( 5 , 0:0.05:2pi ) )' .+ 10
plot( [coords1[:,1],coords2[:,1]] , [coords1[:,2],coords2[:,2]] )
```
