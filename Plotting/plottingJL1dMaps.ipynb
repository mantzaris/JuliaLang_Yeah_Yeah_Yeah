### <span style="color:orange">What are **shape files (.shp)**?</span>

Geographic features in a shapefile can be represented by points, lines, or polygons (areas). It is a common spatial data format developed by ESRI. (from wikipedia) 'The shapefile format can spatially describe vector features: points, lines, and polygons, representing, for example, water wells, rivers, and lakes. (Filename extensions:	.shp, .shx, .dbf)'
This format consists of 3 primary files with the same name, but the following file extensions:

- .shp - shape format; the feature geometry itself. ( main file (.shp) contains the geometry data)
- .dbf - shape index format; a positional index of the feature geometry to allow seeking forwards and backwards quickly (stores the attributes for each shape)
- .shx - attribute format; columnar attributes for each shape, in dBase IV format (the index contains positional index of the feature geometry)

The term "shapefile" is quite common, but the format consists of a collection of files with a common filename prefix, stored in the same directory. The three mandatory files have filename extensions .shp, .shx, and .dbf. The actual shapefile relates specifically to the .shp file, but alone is incomplete for distribution as the other supporting files are required.


From https://www.gislounge.com/what-is-a-shapefile/: *Despite its name indicating a singular file, a shapefile is actually a collection of at least three basic files: .shp, .shx and .dbf.  All three files must be present in the same directory for them to be viewable.  There may be additional files such as a .prj with the shapefileâ€™s projection information.  Commonly, shapefiles are compressed in a .zip file for transfer such as emailing as an attachment or via a web site download. All shapefile files must have the same name but have different extensions.  There are three required files that, at a minimum, make up a shapefile.  For example, a water GIS data set must have the minimum set of files: water.shp, water.shx, and water.dbf*

```julia
using Plots
using Shapefile
using RecipesBase
```

```julia
@recipe function f(sf::Shapefile.Handle)
    # Extract coordinates and shapes
    xs = Float64[]
    ys = Float64[]
    shapes = Shapefile.shapes(sf)

    for shape in shapes
        append!(xs, [point.x for point in shape.points])
        append!(ys, [point.y for point in shape.points])
        append!(xs, NaN)
        append!(ys, NaN)
    end

    # Set plot attributes
    seriestype := :shape
    fillcolor := :yellow
    fillrange := 0
    linecolor := :black

    # Return coordinates
    xs, ys
end
```

```julia
# Load the shapefile
florida_shapefile_path = file_SHP
florida_sf = Shapefile.open(florida_shapefile_path)

# Plot the map
plot(florida_sf)
```
