```julia
using Plots, Distributions
```

```julia
er = Normal(0,5)
```

```julia
x = 1:100
y = collect( x )
y_er = y .+ rand( er, 100 )
```

```julia
plot( x , y , label="no noise" )
plot!( x , y_er , seriestype=:scatter , label="with noise" )
```

```julia
#save the figure
savefig( "testplot.png" )
savefig( "testplot.pdf" )
```

```julia
plot( x , y , label="no noise" )
plot!( x , y_er , seriestype=:scatter , label="with noise" )
plot!( legend=:outerbottom, legendcolumns=2 )
```

```julia
plot( x , y , label="no noise" )
plot!( x , y_er , seriestype=:scatter , label="with noise" )
plot!( legend=:outerbottom, legendrows=2 )
```

```julia
plot( x , y , label="no noise" , linewidth = 10 )
plot!( x , y_er , seriestype=:scatter , label="with noise" )
p1 = plot!( legend=:outertop, legendrows=2 )

plot( x , y , label="no noise" , linewidth = 10 )
plot!( x , y_er , seriestype=:scatter , label="with noise" )
p2 = plot!( legend=:outerright, legendrows=2 )

plot( x , y , label="no noise" , linewidth = 10 )
plot!( x , y_er , seriestype=:scatter , label="with noise" )
p3 = plot!( legend=:outerbottom, legendrows=2 )

plot( x , y , label="no noise" , linewidth = 10 )
plot!( x , y_er , seriestype=:scatter , label="with noise" )
p4 = plot!( legend=:outerleft, legendrows=2 )

plot( p1,p2,p3,p4 , layout=grid(2,2) )
```

```julia
function circle( r , angle ) 
    x = r * cos( angle )
    y = r * sin( angle )
    return [x,y]
end
```

```julia
circle( 1 , 2pi/8 )
```

```julia
angles = 0:0.05:2pi
```

```julia
coords = circle.( 2 , angles)
coords = reduce(hcat,coords)'
```

```julia
plot( coords[:,1] , coords[:,2] )
```

```julia
coords1 = reduce(hcat, circle.( 2 , 0:0.05:2pi ) )'
coords2 = reduce(hcat, circle.( 5 , 0:0.05:2pi ) )' .+ 10
plot( [coords1[:,1],coords2[:,1]] , [coords1[:,2],coords2[:,2]] )
```

```julia
#change the size
plot( [coords1[:,1],coords2[:,1]] , [coords1[:,2],coords2[:,2]] , size = (200,200) )
```

```julia
#set histogram bins
data = rand(100).*10 .+ 150
h1 = histogram( data , title="auto bins" )
h2 = histogram( data , bins = range( start=minimum(data) , stop=maximum(data) , length=20 ) , title="manual bins" )
plot( h1 , h2 )
```

```julia
#normalize histogram
data = rand(100).*10 .+ 150
h1 = plot( data , normalize=true , seriestype=:histogram , title="auto bins" )
h2 = plot( data , normalize=true , seriestype=:histogram , bins = range( start=minimum(data) , stop=maximum(data) , length=20 ) , title="manual bins" )
plot( h1 , h2 )
```

```julia
#change orientation
data = rand(100).*10 .+ 150
h1 = plot( data , orientation=:horizontal , normalize=true , seriestype=:histogram , title="auto bins" )
h2 = plot( data , orientation=:horizontal , normalize=true , seriestype=:histogram , bins = range( start=minimum(data) , stop=maximum(data) , length=20 ) , title="manual bins" )
plot( h1 , h2 )
```

```julia
#ribbon plot
lower = LinRange(0,8,11)
higher = LinRange(0,12,11)
plot( 0:10 , 0:10 , ribbon = (lower,higher) )
```

```julia
labels_tmp = [ "ok" for _ in 1:length(0:10) ]
labels_tmp[end-1] = "UnsureZone"
labels_tmp[[1,end-2,end]] .= ""
plot( 0:10 , 0:10 , ribbon = (lower,higher) , series_annotations = labels_tmp )
```
