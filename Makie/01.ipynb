### Makie is a plotting ecosystem and similar to Plots it can take various backends

The possible backends at this time of writing are:
- CairoMakie.jl (good for vector graphics without interactivity)
- GLMkakie for interactivity and rendering without vector output
- WGLMakie for web browser views

Keywords are *Figure* (containing a *Axis* object). The Axis contains objects such as *Lines*, *Scatter* etc. Think of the *Figure* as container for the other objects, and like a canvas which other objects are added to like the Lines, Legends, Axis etc.

## the Figure and Axis objects were *explicitly* created but they can be created as a byproduct as well

- **lines!** is to add onto an Axis object that already exists
- **lines** (without the exclamation mark) creates a new axis to plot *implicitly*



```julia
xx = range(start=5,stop=12,length=14)
yy = 2 .* xx .^2

lines( xx, yy )
```

```julia
xx = range(start=5,stop=12,length=14)
yy = 2 .* xx .^2

fig4, axis4, lineplot4 = lines( xx, yy )
```

```julia
xx = range(start=5,stop=12,length=14)
yy = 2 .* xx .^2

#the extra comma is needed to declare a tuple
fig4, axis4, lineplot4 = lines( xx, yy, figure= (resolution=(300,300),) )
```

```julia
xx = range(start=5,stop=12,length=14)
yy = 2 .* xx .^2

#the extra comma is needed to declare a tuple
fig4, axis4, lineplot4 = lines( xx, yy, 
                                figure= (resolution=(300,300),),
                                axis=(title="great title",xlabel="x data",ylabel="y data") )
```
