```julia
using HTTP, JSON3
```

```julia
#SWAPI, start wars API
```

```julia
url = "https://swapi.dev/api/people/1/"
response = HTTP.get(url)
character = JSON3.read(response.body)

# Display some details
println("Name: ", character.name)
println("Height: ", character.height, " cm")
println("Mass: ", character.mass, " kg")
println("Hair color: ", character.hair_color)
```

```julia
function fetch_all_people()
    people = []
    next_url = "https://swapi.dev/api/people/"
    
    while !isnothing(next_url)
        response = HTTP.get(next_url)
        data = JSON3.read(response.body)
        append!(people, data.results)
        next_url = data.next  # Get the URL for the next page of results
    end
    
    return people
end

# Fetch all people
all_people = fetch_all_people()

#number of characters per species
species_count = Dict()
for person in all_people
    species = isempty(person.species) ? "Unknown" : basename(person.species[1])
    species_count[species] = get(species_count, species, 0) + 1
end

#  the count of characters in each species
for (spec, count) in species_count
    println("$spec: $count")
end
```

```julia
# PokeAPI
```

```julia
url = "https://pokeapi.co/api/v2/pokemon/pikachu"
response = HTTP.get(url)
pikachu = JSON3.read(response.body)

# Display some details about Pikachu
println("Name: ", pikachu.name)
println("ID: ", pikachu.id)
println("Height: ", pikachu.height, " (decimetres)")
println("Weight: ", pikachu.weight, " (hectograms)")
println("Types:")
for t in pikachu.types
    println("  - ", t.type.name)
end
```

```julia
function fetch_pokemon_by_type(type_name)
    url = "https://pokeapi.co/api/v2/type/" * type_name
    response = HTTP.get(url)
    type_data = JSON3.read(response.body)

    println("Pokémon of type ", type_name, ":")
    for pokemon in type_data.pokemon
        println("  - ", pokemon.pokemon.name)
    end
end

# Fetch and list all fire-type Pokémon
fetch_pokemon_by_type("fire")
```

```julia
country_code = "US"  #country code for the United States
indicator_code = "NY.GDP.MKTP.CD"  #indicator code for GDP
url = "http://api.worldbank.org/v2/country/$(country_code)/indicator/$(indicator_code)?format=json"


response = HTTP.get(url)
data = JSON3.read(response.body)

#data is in the second element of the returned array
if length(data) > 1
    gdp_data = data[2]
    for entry in gdp_data
        if !isnothing(entry.value)
            println("Year: ", entry.date, " - GDP: \$", entry.value)
        end
    end
else
    println("No data available")
end
```

```julia

#  the URL for fetching life expectancy data for Japan
country_code = "JP"  #country code for Japan
indicator_code = "SP.DYN.LE00.IN"  # idicator code for life expectancy at birth
url = "http://api.worldbank.org/v2/country/$(country_code)/indicator/$(indicator_code)?format=json&per_page=60"  # Increase per_page if you need more entries


response = HTTP.get(url)
data = JSON3.read(response.body)


if length(data) > 1 && !isempty(data[2])
    life_expectancy_data = data[2]
    println("Life Expectancy at Birth in Japan (Years):")
    for entry in life_expectancy_data
        year = entry.date
        value = entry.value
        if !isnothing(value)
            println("Year: $year - Life Expectancy: $value")
        end
    end
else
    println("No data available.")
end
```
