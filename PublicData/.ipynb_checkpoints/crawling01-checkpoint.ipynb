```julia
url = "https://www.accuweather.com/en/us/fl/florida-weather"
```

```julia
using HTTP
using Cascadia
using Gumbo
```

```julia
res = HTTP.get(url);
```

```julia
parsed_html = parsehtml(String(res.body))
```

```julia
# make a selector for the element with class: 'text temp'
selector = Selector(".text.temp")
```

```julia
temp_element = eachmatch(selector, parsed_html.root)
```

```julia
temp_text = isempty(temp_element) ? "Element not found" : nodeText(temp_element[1])
```

```julia
#now a selector for elements with class 'text title no-wrap'
selector = Selector(".text.title.no-wrap")

#elements matching that selector
elements = eachmatch(selector, parsed_html.root)
```

```julia
#get the elements for 'Boca Raton'
boca_raton_elements = filter(el -> occursin("Boca Raton", nodeText(el)), elements)
```

```julia
function element_has_class(element, class_name)
    for attribute in element.attributes

        if attribute.first == "class" && occursin(class_name, attribute.second)
            return true
        end
    end
    return false
end

function find_next_sibling_with_class(element, class_name)
    parent = element.parent
    found_current = false
    for child in parent.children
        if found_current && isa(child, Gumbo.HTMLElement) && element_has_class(child, class_name)
            return child
        end
        if child == element
            found_current = true
        end
    end
    return nothing  
end


for el in boca_raton_elements
    sibling = find_next_sibling_with_class(el, "text temp")
    if sibling != nothing
        println("Found sibling with class 'text temp': ", nodeText(sibling))
    else
        println("No sibling found with class 'text temp' for the element containing: ", nodeText(el))
    end
end
```

```julia
boca_link = 0;

anchor_tags = eachmatch(Selector("a.nearby-location.weather-card"), parsed_html.root)

for anchor in anchor_tags
    #if any child span contains the text "Boca Raton"
    span_elements = eachmatch(Selector("span.text.title.no-wrap"), anchor)
    for span in span_elements
        if strip(nodeText(span)) == "Boca Raton"
            #boca anchor tag 'a'
            href_value = get_attribute(anchor, "href")
            if href_value != nothing
                println("Extracted href: ", href_value)
                boca_link = href_value
                break
            else
                println("No href attribute found in the anchor tag.")
            end
        end
    end
end

boca_link
```

```julia
boca_link = 0;

function find_nearest_enclosing_anchor(element)
    while element != nothing && !isa(element, Gumbo.NullNode)
        println("Current element: ", string(element))  # Debug output
        if isa(element, Gumbo.HTMLElement) && lowercase(string(element)) == "a"
            return element
        end
        element = element.parent
    end
    return nothing  # Return nothing if no anchor tag is found
end

if !isempty(boca_raton_elements)
    element = boca_raton_elements[1]
    #println("Direct parent: ", string(element.parent))  # Debug output
    
    nearest_anchor = find_nearest_enclosing_anchor(element)

    if nearest_anchor != nothing
        href_value = get_attribute(nearest_anchor, "href")
        if href_value != nothing
            println("Extracted href: ", href_value)
            boca_link = href_value
        else
            println("No href attribute found in the anchor tag.")
        end
    else
        println("No enclosing anchor tag found.")
    end
else
    println("boca_raton_elements is empty.")
end

boca_link
```
