### <span style="color:orange">Different ranges can be produced by using the keyword arguments</span>

**range(start[, stop]; length, stop, step=1)**

- start (non-optional)
- stop
- length
- step 

all of these do not need to be supplied in each function call, there is a logical reasoning which can be deduced

````julia
#ERROR: without specifying the second value name
range(1,10)
````

````julia
#step-size of '1' is the default if not specified
range(1,stop=10)
````

````julia
#step-size of '0.5'
x7_step_pnt5 = range(1,stop=10,step=0.5)
````

````julia
#length of the range of values
length(x7_step_pnt5)
````

````julia
#y points for these x axis values
y7_step = rand(length(x7_step_pnt5))
````

````julia
#see the increments are at 0.5
plot(x7_step_pnt5,y7_step)
````

````julia
#Not specifying the stopping point but only the step size and the length as the number of points
x8_step_pnt75 = range(1,length=40,step=0.75)
y8 = rand(length(x8_step_pnt75))
````

````julia
plot(x8_step_pnt75,y8)
````

### <span style="color:orange">When you want to call a function with an array of values ( multiple values independently called within an array) you can use either an array comprehension (similar to the list comprehension in python) or the *vectorized* Julia operator '.' (dot) that is a broadcast</span>

````julia
#array comprehension way
[sin(x) for x in [1,12,18,82]]
````

````julia
#or use the convenient broadcast operator to vectorize the computation
sin.([1,12,18,82])
````

````julia
#Now we can specify the end stop point and the length so that the step size is computed
x9_10radiancycles = range(0,stop=10*(2*pi),length=200) #10 full cycles in radians
y9 = sin.(x9_10radiancycles)
````

````julia
plot(x9_10radiancycles,y9)
````

### <span style="color:orange">Within the plot function we can directly call the compute functions for the values we want and put inside the ranges</span>

````julia
#avoids creating the intermediate variables
plot( range(0,stop=10*(2*pi),length=200),sin.(x9_10radiancycles) )
````

### <span style="color:orange">You can pass an **anonymous function**. If this is new to you, don't worry, it will grow on you as a concept. You can write a function which is unnamed, cannot be referenced outside its use, so that instead of calling a function like 'sin' or 'cos' that produce trigonomentric numbers, it gives a specified computation that can even include those functions as well. It makes code more succinct and easy to follow once you get over the hurdle of using them</span>

eg
- x -> x/60
- x -> sin(x/3)
- x -> 1 - x/60

````julia
plot( range(0,stop=10*(2*pi),length=200), x->x/60)
````

````julia
plot( range(0,stop=10*(2*pi),length=200), [sin, cos, x->x/60])
````

````julia
plot( range(0,stop=10*(2*pi),length=200), [sin, cos, x->sin(x/3)])
````

````julia
plot( range(0,stop=10*(2*pi),length=200), [sin, cos, x->(1-x/60)])
````

### <span style="color:orange">Creating functions is really straight forward and there are not too many things to consider when making them, eg:</span>

```
function f1(param1, param2)
tmpVar = param1 + param2
return tmpVar
end
```

or 

```
f2(param1,param2) = return param1+param2;
```

````julia
#example inline function to square the value
f2(param1) = return param1^2;
````

````julia
#call the new function
f2(4)
````

````julia
##try calling it using an array
f2([1,2,3,4])
````

````julia
##try calling it using an array USING BROADCAST
f2.([1,2,3,4])
````

````julia
#use the new user defined function to define the plot mapping of values
plot( range(0,stop=10*(2*pi),length=200), [f2])
````

````julia
#use the modifier functions to add the limits after the plot is created
NN = 20 
varNum = 2
plot( 1:NN, rand(NN,varNum), title="$(NN) random numbers of $(varNum) sequences", xlabel="NN",ylabel="random values")
xlims!(1-2,NN+2)
ylims!(0-1,1+1)                      
````

````julia
#tell plots what the xticks and yticks should be
NN = 20 
varNum = 2
plot( 1:NN, rand(NN,varNum), title="$(NN) random numbers of $(varNum) sequences", xlabel="NN",ylabel="random values",
          xticks=0:2:NN,yticks=[0,0.5,1])
````

````julia
#add different labels to the ticks on the y-axis
NN = 20 
varNum = 2
plot( 1:NN, rand(NN,varNum), title="$(NN) random numbers of $(varNum) sequences", xlabel="NN",ylabel="random values",
          xticks=0:2:NN, yticks=( [0,0.5,1], ["low" "medium" "high"] ) )
````

### <span style="color:orange">Take a Range, collect it into an array, and then make it into an array of strings to alter and then feed to become the xtick labels</span>

````julia
xRange = 0:4:NN
xRange_str = string.(xRange)
````

````julia
#alter an entry 
xRange_str[3] = "$(xRange[3]) special"
xRange_str[5] = "$(xRange[5]) cool"
xRange_str
````

````julia
#put in the new labels and the new size on the x axis
NN = 20 
varNum = 2
plot( 1:NN, rand(NN,varNum), title="$(NN) random numbers of $(varNum) sequences", xlabel="NN",ylabel="random values",
          xticks=(xRange, xRange_str), xtickfontsize=11, yticks=( [0,0.5,1], ["low" "medium" "high"] ),ytickfontsize=14 )
````

````julia
#make the title and the label fonts bigger
NN = 20 
varNum = 2
plot( 1:NN, rand(NN,varNum), title="$(NN) random numbers of $(varNum) sequences", xlabel="NN",ylabel="random values",
          xticks=(xRange, xRange_str), xtickfontsize=11, yticks=( [0,0.5,1], ["low" "medium" "high"] ),ytickfontsize=14,
          titlefontsize=18,xguidefontsize=16,yguidefontsize=16)
````
