```r
julia_library("Plots")
```

```r
#now plot 20 random numbers
```

```r
julia_command( "plot(rand(20))" )
```

```r
#add a title
julia_command( 'plot(rand(20),title="some random numbers \n for fun")' )
```

```r
#change the line width
julia_command( 'plot(rand(20),title="some random numbers \n for fun",linewidth=5)' )
```

```r
#add some markers
julia_command( 'plot(rand(20),title="some random numbers \n for fun",linewidth=5,markershape = :circle)' )
```

```r
#increase marker size
julia_command( ' plot( rand(20), title = "some random numbers \n for fun", linewidth = 5, markershape = :circle, markersize = 8 ) ' )
```

```r
#increase title font size
julia_command( ' plot( rand(20), title = "some random numbers \n for fun", linewidth = 5, markershape = :circle, markersize = 8, titlefontsize = 20 ) ' )
```

```r
#title for the legend
julia_command( ' plot( rand(20), title = "some random numbers \n for fun", linewidth = 5, markershape = :circle, markersize = 8, titlefontsize = 20, legendtitle = "line names" ) ' )
```

```r
#change line name
julia_command( ' plot( rand(20), title = "some random numbers \n for fun", linewidth = 5, markershape = :circle, markersize = 8, titlefontsize = 20, legendtitle = "line names", label = "uniform random" ) ' )
```

```r
#change tick font size
julia_command( ' plot( rand(20), title = "some random numbers \n for fun", linewidth = 5, markershape = :circle, markersize = 8, titlefontsize = 20, legendtitle = "line names", label = "uniform random", xtickfontsize = 14, ytickfontsize = 14 ) ' )
```

```r
#now put 4 independent variables to plot
julia_command( ' plot( rand(20,4), title = "some random numbers \n for fun" ) ' )
```

```r
#how do we get some important R data into Julia?
Rtmp = c(1,22,333,4444)
Jtmp = JuliaObject( Rtmp )
Jtmp
```

```r
julia_assign("x2",Jtmp)
```

```r
julia_eval("x2")
```
