## Cross entropy

We have 2 probability distributions $p$ and $q$ for the same *domain*. Here $p$ is the true distribution, and $q$ an approximation. In terms of information theory it measures the average (expected) number of bits to code one in terms of the other. In the discrete case it is

$H(p,q) = - \sum_{x \in domain} p(x) log(q(x))$

often we do not know $p(x)$ beforehand. This is related to the KL divergence measure (distance)

$D_{KL}(P||Q) = \sum_{x\in domain} P(x) log( \frac{P(x)}{Q(x)} )$



```julia
using Flux
using Flux:onehotbatch
using Flux:crossentropy,onecold,throttle
using Random
using Statistics:mean
using IterTools
```

```julia
Fs = Fs_mechanical_test[1]
max_freq = 10^4
function Get_Powers_freq10K(signal)
    prdgrm_tmp = DSP.Periodograms.periodogram(signal, fs=Int(Fs), window=DSP.Windows.hamming)
    original = prdgrm_tmp.power[1:max_freq]
    return original ./ sum(original)
end
```

```julia
p1 = plot( Get_Powers_freq10K(mechanical_signals[1]) )
p2 = plot( Get_Powers_freq10K(crowd_signals[1]) )
plot( p1 , p2 )
```

```julia
#make into a set of columns of arrays of floats (rows:dim of data by columns: number of samples)
X_tmp = vcat( Get_Powers_freq10K.(crowd_signals) , Get_Powers_freq10K.(mechanical_signals) )
X = Flux.stack( X_tmp )
X = Flux.unsqueeze( X , dims = 2 )
println( size(X) )
println( typeof(X) )
```

```julia
#use onehotbatch by providing it the labels (target value correct) and the range of targets it can take
Y = [ repeat( ["crowd"] , length(crowd_signals) ) ; repeat( ["mechanical"] , length(mechanical_signals) ) ]
```

```julia
Y_1hot = Flux.onehotbatch(Y,["crowd","mechanical"])
```

```julia
display( size( X ) )
#model = Chain( Conv((5,),1=>2) , Conv((5,),2=>2) , AdaptiveMeanPool( (100,) ) ,
#               Conv((3,),2=>1,relu,pad=SamePad()) , Flux.flatten , Dense(100=>20,relu), Dense(20=>2,relu), softmax )

model = Chain( Flux.flatten , Dense(10000=>2,relu), softmax )
```

```julia
model( X )
```

```julia
size( model( X ) )
```

```julia
#define the loss function; eg: loss(X, y) = crossentropy(prediction value, target value) 
loss1(m,X,y) = Flux.crossentropy(m(X),y)
#set the optimization function
opt = Flux.setup( Flux.Adam(0.001) , model );
```

```julia
loss1( model, X , Y_1hot )
```

```julia
data = ncycle( [(X,Y_1hot)] , 20 )
data_sound = Flux.DataLoader( (X,Y_1hot) , batchsize=10 , shuffle=true ) 
```

```julia
Flux.onecold( model( X ) , ["crowd","mechanical"] )
```

```julia
#just a print out macro for the loss function we made
losses = []

for epoch in 1:10000
    
    for ( x , y ) in data_sound    
        
        loss_tmp, grads = Flux.withgradient( model ) do model
                
            y_hat = model( x )
            loss = Flux.crossentropy( y_hat , y )
            return loss
            
        end     
        Flux.update!( opt , model , grads[1] )
        push!(losses,loss_tmp) #don't put this in the above block where the gradient is computed
    end
#     break
end
p1 = plot( losses ,  legend=false , size=(600,150) )
display(p1)
println(model( X ))
```

```julia
Flux.onecold( model( X ) , ["crowd","mechanical"] )
```

```julia
println( "accuracy = ",sum( Flux.onecold( model( X ) , ["crowd","mechanical"] ) .== Flux.onecold( Y_1hot , ["crowd","mechanical"] ) )*100 ./ length(Y) )
```
