# Q1

data from: https://www.kaggle.com/datasets/joebeachcapital/banana-index

Q1a

- Load the data, and remove unnecessary columns and missing/NA
- select a subset of 2 columns useful for clustering
- produce a scatter plot of the data, and then a separate scatter plot of the log of the data variables, discuss what you should use for clustering and the differences

```julia
banana_data = CSV.File("./bananaindex.csv") |> DataFrame;
println(size(banana_data))
select!(banana_data, Not([Symbol("Banana values"), Symbol("Unnamed: 16"), Symbol("Chart?"),:type,:year]))
names(banana_data)
```

```julia
dropmissing!(banana_data);
println(size(banana_data))
```

```julia
banana_data_subset = Matrix(banana_data[:, [Symbol("Bananas index (kg)"), Symbol("land_use_1000kcal")]]);
```

```julia
banana_data_subset_log = log.(10,banana_data_subset);
```

```julia
p1 = plot(banana_data_subset[:,1], banana_data_subset[:,2], seriestype=:scatter, legend=false, 
        title="Banana index (kg) and land_use_1000kcal", titlefontsize=10, margin=3mm)
xlabel!("Banana index (kg)")
ylabel!("land_use_1000kcal")
p2 = plot(banana_data_subset_log[:,1], banana_data_subset_log[:,2], seriestype=:scatter, legend=false,
        title="Banana index (kg) and land_use_1000kcal \n log10", titlefontsize=10, margin=3mm)
xlabel!("log Banana index (kg)")
ylabel!("log land_use_1000kcal")
p = plot(p1,p2, layout=grid(1,2), size=(700,300))
display(p)
```

### Q1b 

- perform k-medoids clustering and choose a suitable number for K and discuss why (explore the key change for different k).

```julia
D = pairwise(Euclidean(), banana_data_subset_log', banana_data_subset_log');
```

```julia
size(D)
```

```julia
total_costs = [ kmedoids(D, k).totalcost for k in 1:8 ];
```

```julia
p = plot( total_costs, title="total costs for k in k-medoids", linewidth=6, size=(600,300) )
display(p)
```

### Q1c

- using your choice of 'k', plot the points with a color for each cluster label and place a marker in the center of each cluster, and the name of the food for that medoid.
- looking at the separation of the foods based on clusters,discuss any major differences in the cluster assignments

```julia
k=2
result_k2 = kmedoids(D, k);
centers_k2_inds = result_k2.medoids;
centers_k2 = banana_data_subset_log[centers_k2_inds,:]
medoids_food_labels = banana_data[centers_k2_inds,1]
```

```julia
p1 = plot(banana_data_subset_log[:,1], banana_data_subset_log[:,2], seriestype=:scatter, legend=false,
        title="Banana index (kg) and land_use_1000kcal \n log10, with cluster assignments", titlefontsize=12,
        marker_z=result_k2.assignments, color=:lightrainbow, margin=3mm )

plot!(centers_k2[:,1], centers_k2[:,2], marker=[:star, :circle], color=[:black, :blue], markersize=8,
            label="Cluster Centers", seriestype=:scatter, margin=3mm )

for row_num in 1:size(centers_k2,1)
    annotate!( centers_k2[row_num,1], centers_k2[row_num,2] , text(medoids_food_labels[row_num] ,"Bookman Demi",:black, :top, 17) )
end

xlabel!("log Banana index (kg)")
ylabel!("log land_use_1000kcal")
p = plot(p1, size=(700,300))
savefig("fig3.png")
display(p)
```

### Q1d

- For each point find the distance to each medoid (cluster center which is an exemplar) for each point (food). This should produce 2 distance values. Produce the difference between those distances and display the set on a histogram.
- Find the 3 closest to zero on that distribution and comment on the foods selected. Plot those foods as annotations on the scatter plot and discuss the findings for any pattern which can be speculated.

```julia
D = pairwise( Euclidean() , banana_data_subset_log' , centers_k2' );
diff_medoids = map( row ->  abs(row[1] - row[2])  , eachrow(D) );
```

```julia
plot( diff_medoids, title="differences between distances to medoids" , seriestype=:histogram, legend=false,
            bins=range(minimum(diff_medoids),stop=maximum(diff_medoids),length=15 ) )
```

```julia
smallest_two = sortperm(diff_medoids)[1:2]
closest_to_zero = sortperm(diff_medoids, by=abs)[1:3]
largest_two = sortperm(diff_medoids, rev=true)[1:2]
```

```julia
banana_data[ closest_to_zero ,1]
```

```julia
border_inds = [closest_to_zero...];
```

```julia
annotation_text_tmp = banana_data[ border_inds , 1 ]
```

```julia
p1 = plot(banana_data_subset_log[:,1], banana_data_subset_log[:,2], seriestype=:scatter, legend=false,
        title="Banana index (kg) and land_use_1000kcal \n log10, with cluster assignments", titlefontsize=12,
        marker_z=result_k2.assignments, color=:lightrainbow )

plot!(centers_k2[:,1], centers_k2[:,2], marker=[:circle, :circle], color=[:black, :black], markersize=8,
            label="Cluster Centers", seriestype=:scatter )


for ind in 1:length(border_inds)
    
     plot!( [banana_data_subset_log[border_inds[ind],1]], [banana_data_subset_log[border_inds[ind],2]] ,
                    marker=:star, color=:blue, markersize=8,
                    seriestype=:scatter )
    
    annotate!( banana_data_subset_log[border_inds[ind],1], banana_data_subset_log[border_inds[ind],2] , 
                                                                text(annotation_text_tmp[ind] ,:blue, :top, 12) )
    
end

xlabel!("log Banana index (kg)")
ylabel!("log land_use_1000kcal")
p = plot(p1, size=(700,300), margin=3mm)
savefig("fig4.png")
display(p)
```

### Q1e

- Find the 6 most extreme of the foods and discuss any pattern plotting the names as an annotation

```julia
far_medoids = map( row ->  abs(row[1] - row[2]) * (row[1] + row[2])  , eachrow(D) );
```

```julia
largest = sortperm(far_medoids, rev=true)[1:6]
annotation_text_tmp = banana_data[ largest , 1 ]
```

```julia
p1 = plot(banana_data_subset_log[:,1], banana_data_subset_log[:,2], seriestype=:scatter, legend=false,
        title="Banana index (kg) and land_use_1000kcal \n log10, with cluster assignments", titlefontsize=12,
        marker_z=result_k2.assignments, color=:lightrainbow )

plot!(centers_k2[:,1], centers_k2[:,2], marker=[:circle, :circle], color=[:black, :black], markersize=8,
            label="Cluster Centers", seriestype=:scatter )


for ind in 1:length(largest)
    
     plot!( [banana_data_subset_log[largest[ind],1]], [banana_data_subset_log[largest[ind],2]] ,
                    marker=:star, color=:blue, markersize=8,
                    seriestype=:scatter )
    
    annotate!( banana_data_subset_log[largest[ind],1], banana_data_subset_log[largest[ind],2] , 
                                                                text(annotation_text_tmp[ind] ,:blue, :top, 12) )
    
end

xlabel!("log Banana index (kg)")
ylabel!("log land_use_1000kcal")
p = plot(p1, size=(700,300), margin=3mm)
savefig("fig5.png")
display(p)
```
