```julia
function circle( r , angle ) 
    x = r * cos( angle )
    y = r * sin( angle )
    return [x,y]
end
```

```julia
circle( 1 , 2pi/8 )
```

```julia
angles = 0:0.05:2pi
```

```julia
coords = circle.( 2 , angles)
coords = reduce(hcat,coords)'
```

```julia
plot( coords[:,1] , coords[:,2] )
```

```julia
#set histogram bins
data = rand(100).*10 .+ 150
h1 = histogram( data , title="auto bins" )
h2 = histogram( data , bins = range( start=minimum(data) , stop=maximum(data) , length=20 ) , title="manual bins" )
plot( h1 , h2 )
```

```julia
#normalize histogram
data = rand(100).*10 .+ 150
h1 = plot( data , normalize=true , seriestype=:histogram , title="auto bins" )
h2 = plot( data , normalize=true , seriestype=:histogram , bins = range( start=minimum(data) , stop=maximum(data) , length=20 ) , title="manual bins" )
plot( h1 , h2 )
```

```julia
#change orientation
data = rand(100).*10 .+ 150
h1 = plot( data , orientation=:horizontal , normalize=true , seriestype=:histogram , title="auto bins" )
h2 = plot( data , orientation=:horizontal , normalize=true , seriestype=:histogram , bins = range( start=minimum(data) , stop=maximum(data) , length=20 ) , title="manual bins" )
plot( h1 , h2 )
```

```julia
#Animations with the @animate and @gif macros

x = 1:20

animation1 = @animate for ii in 1:10
    heatmap( x , x , hm1 , c = :cool , colorbar = true )
end
gif( animation1 , "testAnimation.gif" , fps = 2 )
```
