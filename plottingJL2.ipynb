```julia
#make a 'noisy' square
function noisySquare()
    newPic = ones(NN,NN)
    for rr = 1:NN
        for cc = 1:NN
            if( (rr >= 3 && rr <= 7) && (cc >= 3 && cc <= 7) )
                if( rand() >= 0.2 ) newPic[rr,cc] = 0 end
            end
        end
    end
    return newPic
end
```

```julia
#function 'noisy cross'
function noisyCross()
    newPic = ones(NN,NN)
    for rr = 1:NN
        for cc = 1:NN
            if( rr == 5 || cc == 5 )
                if( rand() >= 0.2 ) newPic[rr,cc] = 0 end
            end
        end
    end
    return newPic
end
```

```julia
#make a triangle
trianglePic = ones(NN,NN)
trianglePic[5,5] = 0
trianglePic[5,4] = 0
trianglePic[5,3] = 0
trianglePic[4,5] = 0
trianglePic[4,4] = 0
trianglePic[3,5] = 0
trianglePic[5,6] = 0
trianglePic[5,7] = 0
trianglePic[4,6] = 0
trianglePic[4,5] = 0

Gray.(trianglePic)
```

```julia
#make a 'noisy' Triangle
function noisyTriangle()
    newPic = copy(trianglePic)
    for rr = 1:NN
        for cc = 1:NN
            if( rand() <= 0.2 ) 
                newPic[rr,cc] = 1
            end
        end
    end
    return newPic
end
```

### <span style="color:orange">Now insert a common component of variance between classes that exists *within* each class</span>

```julia
function noisySquare2()
    newPic = ones(NN,NN)
    for rr = 1:NN
        for cc = 1:NN
            if( (rr >= 3 && rr <= 7) && (cc >= 3 && cc <= 7) )
                if( rand() >= 0.2 ) newPic[rr,cc] = 0 end
            end
        end
    end
    if( rand() >= 0.5 )
         newPic[NN-2:NN,NN-2:NN] .= 0     
    end
    return newPic
end
```

```julia
function noisyCross2()
    newPic = ones(NN,NN)
    for rr = 1:NN
        for cc = 1:NN
            if( rr == 5 || cc == 5 )
                if( rand() >= 0.2 ) newPic[rr,cc] = 0 end
            end
        end
    end
    if( rand() >= 0.5 )
         newPic[NN-2:NN,NN-2:NN] .= 0     
    end
    return newPic
end
```
